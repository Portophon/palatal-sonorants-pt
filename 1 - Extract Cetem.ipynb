split -l 10000000 -d --additional-suffix=.txt CETEMPublico1.7 ./cetem_corpus/

```python
import os
import re
import string
import pandas as pd
```

```python
string.punctuation
```

```python
remove_words = string.punctuation
remove_words = remove_words.replace("-", "") # don't remove hyphens
remove_words = remove_words.replace("\\", "") # don't remove slash
remove_words = remove_words + "0123456789ºª«»" 
remove_words
```

```python
lh_est = []
with open("lh_est.txt", encoding="utf-8") as file:
    for line in file:
        line = line.rstrip()
        lh_est.append(line)
print(len(lh_est))
print(lh_est)
```

```python
nh_est = []
with open("nh_est.txt", encoding="utf-8") as file:
    for line in file:
        line = line.rstrip()
        nh_est.append(line)
print(len(nh_est))
print(nh_est)
```

```python
inho_no_dim = []
with open("inhoas_no_dim.txt", encoding="utf-8") as file:
    for line in file:
        line = line.rstrip()
        inho_no_dim.append(line)
inho_no_dim
```

```python
total_words = 0
total_words_lh = 0
total_words_lh_clitic = 0
total_words_mesoclisis = 0
total_words_nh = 0
total_words_diminutive = 0
total_words_p = 0
total_words_m = 0
total_words_m_clitic = 0

path_to_file = "cetem_corpus/"
cetem_corpus = os.listdir(path_to_file)
#cetem_new_corpus = "cetem_p"
cetem_corpus.sort() 

num_words = 0
#word_list_set = []

for file_name in cetem_corpus:
    word_list = []
    #if file_name not in cetem_new_corpus:
    with open(path_to_file + file_name, encoding="latin-1") as file:
        for line in file:
            line = line.rstrip()
            word_list.append(line)
    print(file_name + ": " + str(len(word_list)))            
    
    word_list = [x for x in word_list if not any(p in x for p in remove_words)]
    print("after removing words with tags punctuation and numbers: "  + str(len(word_list)))
    
    word_list = [re.sub("(\-)+$", "", i) for i in word_list]
    
    word_list = [re.sub("(?<=([a-z]|[áéíóúãõâêô]))É$", " É", i) for i in word_list]

    word_list = [re.sub("(\-)(?=[A-Z]|[ÁÉÍÓÚÂÊÔ]|([mt]e|lh[eao]s?|[mtnl]?[oa]s?|vos)$)", "_", i) for i in word_list]
    word_list = [re.sub("(\-)(?=([mt]e|lh[eao]s?|[mtnl]?[oa]s?|vos)\_)", "_", i) for i in word_list]
    word_list = [i.replace('\\', '_') for i in word_list]
    word_list = [re.sub("^\_+", "", i) for i in word_list]
    word_list = [re.sub("\_+$", "", i) for i in word_list]
    word_list = [i.split("_") for i in word_list]
    word_list = [val for sublist in word_list for val in sublist]   
    word_list = [i.lower() for i in word_list]
    
    print("after spliting words with slashs, as well as clitics and capitalized words after hyphen: "  + str(len(word_list)))
    
    word_list = [re.sub("[^\w\-]", "", i) for i in word_list]
    word_list = [re.sub("^\-+", "", i) for i in word_list]
    
    word_list = ' '.join(word_list).split()
    
    mesoclisis_list = ".*\-([mt]e|lh[eao]s?|[mtnl]?[oa]s?|vos)\-.*"
    r = re.compile(mesoclisis_list)
    pt_mesoclisis = list(filter(r.match, word_list))
    
    verb_list = ".*\-(ia|íamos|ão|á|iam|ei|à|ía|emos|eis|ás|êmos|ao|iamos|íam|ias)$"
    r = re.compile(verb_list)
    pt_mesoclisis = list(filter(r.match, word_list))
    
    mesoclisis_lh_list = ".*\-(lh[eao]s?)\-.*"
    r = re.compile(mesoclisis_lh_list)
    pt_lh_mesoclisis = list(filter(r.match, pt_mesoclisis))
    
    mesoclisis_m_list = ".*\-(me|(m[ao]s?))\-.*"
    r = re.compile(mesoclisis_m_list)
    pt_m_mesoclisis = list(filter(r.match, pt_mesoclisis))
   
    word_list = [re.sub("\-((m|(lh))[eao]s?)\-", "-", i) for i in word_list]
    
    pt_lh = [match for match in word_list if "lh" in match]
    pt_lh = [word for word in pt_lh if word not in lh_est]
    
    clitic_list = "^lh[eao]s?$" 
    r = re.compile(clitic_list)
    pt_lh_clitic = list(filter(r.match, pt_lh))
        
    pt_nh = [match for match in word_list if "nh" in match]
    pt_nh = [word for word in pt_nh if word not in nh_est]
    
    inho_list = ".*(inho|inha|inhos|inhas)$"
    r = re.compile(inho_list)
    pt_diminutive = list(filter(r.match, pt_nh))
    pt_diminutive = [word for word in pt_diminutive if word not in inho_no_dim]
    
    pt_p = [match for match in word_list if "p" in match]

    m_list = ".*m[aáàâãeéèêiíìoóòôõuúù].*" 
    r = re.compile(m_list)
    pt_m = list(filter(r.match, word_list))
    
    clitic_list = "^m[eao]s?$" 
    r = re.compile(clitic_list)
    pt_m_clitic = list(filter(r.match, pt_m))

    num_words_lh = len(pt_lh) + len(pt_lh_mesoclisis)
    num_words_lh_clitic = len(pt_lh_clitic) + len(pt_lh_mesoclisis)
    num_words_mesoclisis = len(pt_mesoclisis)
    num_words_lh_mesoclisis = len(pt_lh_mesoclisis)
    num_words_nh = len(pt_nh)
    num_words_diminutive = len(pt_diminutive)
    num_words_p = len(pt_p)
    num_words_m = len(pt_m) + len(pt_m_mesoclisis)
    num_words_m_clitic = len(pt_m_clitic) + len(pt_m_mesoclisis)
    num_words_m_mesoclisis = len(pt_m_mesoclisis)
    
    total_words = total_words + len(word_list) + num_words_mesoclisis
    total_words_lh = total_words_lh + num_words_lh
    total_words_lh_clitic = total_words_lh_clitic + num_words_lh_clitic
    total_words_mesoclisis = total_words_mesoclisis + num_words_mesoclisis
    total_words_nh = total_words_nh + num_words_nh
    total_words_diminutive = total_words_diminutive + num_words_diminutive
    total_words_p = total_words_p + num_words_p
    total_words_m = total_words_m + num_words_m 
    total_words_m_clitic = total_words_m_clitic + num_words_m_clitic
    
    MyFile=open("cetem_filtered/" + file_name,'w')
    MyList=map(lambda x:x+'\n', word_list)
    MyFile.writelines(MyList)
    MyFile.close()
    
    MyFile=open("cetem_lh/" + file_name,'w')
    MyList=map(lambda x:x+'\n', pt_lh)
    MyFile.writelines(MyList)
    MyFile.close()
    
    MyFile=open("cetem_lh_clitic/" + file_name,'w')
    MyList=map(lambda x:x+'\n', pt_lh_clitic)
    MyFile.writelines(MyList)
    MyFile.close()
    
    MyFile=open("cetem_m_clitic/" + file_name,'w')
    MyList=map(lambda x:x+'\n', pt_m_clitic)
    MyFile.writelines(MyList)
    MyFile.close()

    MyFile=open("cetem_mesoclisis/" + file_name,'w')
    MyList=map(lambda x:x+'\n', pt_mesoclisis)
    MyFile.writelines(MyList)
    MyFile.close()
   
    MyFile=open("cetem_nh/" + file_name,'w')
    MyList=map(lambda x:x+'\n', pt_nh)
    MyFile.writelines(MyList)
    MyFile.close()
    
    MyFile=open("cetem_diminutive/" + file_name,'w')
    MyList=map(lambda x:x+'\n', pt_diminutive)
    MyFile.writelines(MyList)
    MyFile.close()

    MyFile=open("cetem_p/" + file_name,'w')
    MyList=map(lambda x:x+'\n', pt_p)
    MyFile.writelines(MyList)
    MyFile.close()

    MyFile=open("cetem_m/" + file_name,'w')
    MyList=map(lambda x:x+'\n', pt_m)
    MyFile.writelines(MyList)
    MyFile.close()

    print("num_words_lh: " + str(num_words_lh))
    print("num_words_lh_clitic: " + str(num_words_lh_clitic))
    print("num_words_mesoclisis: " + str(num_words_mesoclisis))
    print("num_words_lh_mesoclisis: " + str(num_words_lh_mesoclisis))
    print("num_words_nh: " + str(num_words_nh))
    print("num_words_diminutive: " + str(num_words_diminutive))
    print("num_words_p: " + str(num_words_p))
    print("num_words_m: " + str(num_words_m))
    print("num_words_m_clitic: " + str(num_words_m_clitic))
    print("num_words_m_mesoclisis: " + str(num_words_m_mesoclisis))
     
    with open('CetemPublicoStatsConsonants.txt', 'a', encoding="latin-1") as f:
        f.write('File: ' + file_name)
        f.write(' # lh words: ' + str(num_words_lh))
        f.write(' # lh clitic words: ' + str(num_words_lh_clitic))
        f.write(' # mesoclisis words: ' + str(num_words_mesoclisis))
        f.write(' # nh words: ' + str(num_words_nh))
        f.write(' # diminutive words: ' + str(num_words_diminutive))
        f.write(' # p words: ' + str(num_words_p))
        f.write(' # m words: ' + str(num_words_m))
        f.write(' # m clitic words: ' + str(num_words_m_clitic))
        f.write('\n')

items_list = ["total_words", "total_words_lh", "total_words_lh_clitic", "total_words_nh", "total_words_diminutive", "total_words_p", "total_words_m", "total_words_m_clitic"]
stats = [total_words, total_words_lh, total_words_lh_clitic, total_words_nh, total_words_diminutive, total_words_p, total_words_m, total_words_m_clitic]
stats_df = pd.DataFrame(
    {'items_list': items_list,
     'stats': stats
    })

stats_df.to_csv('stats_df.csv')
```
