```python
import os
import re
import pandas as pd
```

```python
path_to_file = "cetem_filtered/"
cetem_corpus = os.listdir(path_to_file)

word_list_set = []

for file_name in cetem_corpus:
    word_list = []
    with open(path_to_file + file_name, encoding="utf-8") as file:
        for line in file:
            line = line.rstrip()
            word_list.append(line)
        
    word_list_set = list(set(word_list_set + list(set(word_list))))
    
    print(len(word_list_set))

num_word_list_set = len(word_list_set)
```

```python
MyFile=open("word_list_set.txt",'w')
MyList=map(lambda x:x+'\n', word_list_set)
MyFile.writelines(MyList)
MyFile.close()

word_list_set = []
```

```python
path_to_file = "cetem_lh/"
cetem_corpus = os.listdir(path_to_file)

word_list_set = []
word_list_set_clitic = []
word_list_clitic_total = []
clitic_list = "^lhe$|^lhes$|^lho$|^lhos$|^lha$|^lhas$" 
r = re.compile(clitic_list)

for file_name in cetem_corpus:
    word_list = []
    word_list_clitic = []
    with open(path_to_file + file_name, encoding="utf-8") as file:
        for line in file:
            line = line.rstrip()
            word_list.append(line)
        
    word_list_set = list(set(word_list_set + list(set(word_list))))
    print(len(word_list_set))
    
    
    word_list_clitic = [match for match in word_list if "-lh" in match]
    
    word_list_set_clitic = list(set(word_list_set_clitic + list(set(word_list_clitic))))
    word_list_clitic_total.append(word_list_clitic)
    print(len(word_list_set))
    print(len(word_list_set_clitic))
   
clitic_list = "^lhe$|^lhes$|^lho$|^lhos$|^lha$|^lhas$|.*\-lh.+" 
r = re.compile(clitic_list) 
word_list_clitic = list(filter(r.match, word_list_set))
word_list_set_clitic = list(set(word_list_set_clitic + list(set(word_list_clitic))))
print(len(word_list_set))
print(len(word_list_set_clitic))
word_list_clitic_total = [item for sublist in word_list_clitic_total for item in sublist]
print(len(word_list_clitic_total))

num_word_list_set_lh = len(word_list_set)
```

```python
MyFile=open("word_list_set_lh.txt",'w')
MyList=map(lambda x:x+'\n', word_list_set)
MyFile.writelines(MyList)
MyFile.close()

word_list_set = []
```

```python
path_to_file = "cetem_nh/"
cetem_corpus = os.listdir(path_to_file)

word_list_set = []
word_list_set_inho = []
word_list_inho_total = []


for file_name in cetem_corpus:
    word_list = []
    word_list_inho = []
    with open(path_to_file + file_name, encoding="utf-8") as file:
        for line in file:
            line = line.rstrip()
            word_list.append(line)
    
    word_list_set = list(set(word_list_set + list(set(word_list))))
    
    inho_list = ".*(inho|inha|inhos|inhas)$"
    r = re.compile(inho_list)
    word_list_inho = list(filter(r.match, word_list))
    word_list_inho = [word for word in word_list_inho if word not in inho_no_dim]
    word_list_inho_total.append(word_list_inho)

    
    word_list_set_inho = list(set(word_list_set_inho + list(set(word_list_inho))))
    
    print(len(word_list_set))

word_list_inho_total = [item for sublist in word_list_inho_total for item in sublist]
num_word_list_set_nh = len(word_list_set)
num_word_list_set_inho = len(word_list_set_inho)
word_list_set_inho_df = pd.DataFrame(word_list_set_inho, columns=["words"])
word_list_set_inho_df.to_csv('word_list_set_inho.csv')
```
