```python
import os
import re
import pandas as pd
```

```python
path_to_file = "cetem_filtered/"
cetem_corpus = os.listdir(path_to_file)

word_list_set = []

for file_name in cetem_corpus:
    word_list = []
    with open(path_to_file + file_name, encoding="utf-8") as file:
        for line in file:
            line = line.rstrip()
            word_list.append(line)
        
    word_list_set = list(set(word_list_set + list(set(word_list))))
    
    print(len(word_list_set))

num_word_list_set = len(word_list_set)
```

```python
MyFile=open("word_list_set.txt",'w')
MyList=map(lambda x:x+'\n', word_list_set)
MyFile.writelines(MyList)
MyFile.close()

word_list_set = []
```

```python
path_to_file = "cetem_lh/"
cetem_corpus = os.listdir(path_to_file)

word_list_set = []

for file_name in cetem_corpus:
    word_list = []
    with open(path_to_file + file_name, encoding="utf-8") as file:
        for line in file:
            line = line.rstrip()
            word_list.append(line)
        
    word_list_set = list(set(word_list_set + list(set(word_list))))
    print(len(word_list_set))
       
clitic_lh_list = "^lhe$|^lhes$|^lho$|^lhos$|^lha$|^lhas$|.*\-lh.+" 
r = re.compile(clitic_lh_list) 
word_list_lh_clitic = list(filter(r.match, word_list_set))
word_list_set_lh_clitic = list(set(word_list_lh_clitic))
print(len(word_list_set))
print(len(word_list_set_lh_clitic))

num_word_list_set_lh = len(word_list_set)
```

```python
path_to_file = "cetem_mesoclisis/"
cetem_corpus = os.listdir(path_to_file)

word_list_set = []

for file_name in cetem_corpus:
    word_list = []
    with open(path_to_file + file_name, encoding="utf-8") as file:
        for line in file:
            line = line.rstrip()
            word_list.append(line)
        
    word_list_set = list(set(word_list_set + list(set(word_list))))
    print(file_name)
    #print(word_list_set)
    print(len(word_list_set))

num_word_list_set_mesoclisis = len(word_list_set)

mesoclisis_lh_list = ".*\-(lh[eao]s?)\-.*"
r = re.compile(mesoclisis_lh_list)
lh_mesoclisis_set = list(filter(r.match, word_list_set))
num_word_list_set_lh_mesoclisis = len(lh_mesoclisis_set)
print(num_word_list_set_lh_mesoclisis)

clitic_m_list = ".*\-(me|([eao]s?))\-.*" 
r = re.compile(clitic_m_list) 
m_mesoclisis_set = list(filter(r.match, word_list_set))
num_word_list_set_m_mesoclisis = len(m_mesoclisis_set)
print(num_word_list_set_m_mesoclisis)
```
